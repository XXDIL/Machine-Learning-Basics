# Implementing Box-Muller Transformation Algorithm

#### You need to implement the Box-Muller Algorithm and show that the generated samples are indeed normally distributed by generating a plot.

```python
import numpy as np
import matplotlib.pyplot as plt
import math
```

### Two Independent uniform samples. (u1s and u2s)

```python
rng = np.random.RandomState(seed=42)
u1s, u2s = rng.uniform(size=1000), rng.uniform(size=1000)

fig, axs = plt.subplots(2, 1)
axs[0].hist(u1s, density=True, bins=50)
axs[0].set_title('U1')
axs[1].hist(u2s, density=True, bins=50)
axs[1].set_title('U2')
fig.tight_layout(pad=0.5)
```

------
### Apply the inverse CDF of the exponential distribution with λ=1 to u1s 

> Which means just taking the negative log of u1s => -ln(u1s)

### Apply the inverse CDF of the uniform distribution between 0 and 2π to u2s 

> gives the angle of samples => 2π(u2s)

```python
ss = -np.log(u1s)
thetas = 2*math.pi*u2s

plt.hist(ss, bins=50)
plt.title('ss = -ln(u1s)')
plt.show()
```

### Calculate the distance r from origin for each sample

```python
rs = np.sqrt(2*ss)

plt.hist(rs, density=True, bins=50)
plt.title('rs = sqrt(-2ln(u1s))')
plt.show()
```

### Finding x and y is easy as we have r

```python
xs, ys = rs*np.cos(thetas), rs*np.sin(thetas)

plt.scatter(xs, ys)
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```

### The histogram looks like follows

```python
bins = np.linspace(-4, 4, 50)

plt.hist(xs, bins, alpha=0.5, label='xs')
plt.hist(ys, bins, alpha=0.5, label='ys')
plt.legend(loc='upper right')
plt.show()
```
